```python
# Scenario 1: Iris Dataset Analysis
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Generate a pair plot with species-based color coding
sns.pairplot(iris, hue="species", diag_kind="hist")

# Display the plot
plt.suptitle("Pairwise Relationships in Iris Dataset", y=1.02)
plt.show()
```

```python
# Scenario 2: Car Specifications Exploration
import seaborn as sns
import matplotlib.pyplot as plt

# Load the car dataset
cars = sns.load_dataset("mpg")

# Drop rows with missing values
cars = cars.dropna()

# Select relevant features
features = ["horsepower", "weight", "mpg", "origin"]

# Create the pair plot
sns.pairplot(cars[features], hue="origin", diag_kind="hist")

# Display the plot
plt.suptitle("Relationships Between Car Specifications", y=1.02)
plt.show()
```
